```python
import pandas as pd 
import numpy as np
```

## 1. Reshaping by Stacking and Unstacking


```python
# stack : transforming a dataframe into a series 
# unstacking : transforming a multi-index series into dataframe 

df1 = pd.DataFrame(np.random.randint(50,size = (4,3)) , 
                   columns= ['sale1' , 'sale2' , 'sale3'])
df1
```

```python
df2 = pd.DataFrame(['branch1','branch2','branch3','branch4'],
                   columns = ['branch_nbr'])
df2
```

```python
df3 = pd.concat([df2,df1],axis = 1)
df3
```

```python
df4 = df3.set_index('branch_nbr')
df4
```

```python
#cool 
```

```python
ds1 = df4.stack()
ds1
```

```python
ds1.unstack(0) 
#level 0
```

## 2. Reshaping by Melting (Wide to Long )

```python
df4.reset_index(inplace= True)
```

```python
df4.melt(id_vars = 'branch_nbr')
# id_vars means don'r melt branch_nbr
```

```python
df4.set_index('branch_nbr' , inplace = True)
df4
```

```python
# to keep the index 
df5 = df4.melt(var_name = 'sales' , value_name= 'value' ,ignore_index= False)
```

## 3. Reshaping by Pivoting (Long to Wide)

```python
df5
```

```python
df5.pivot( columns = 'sales', values= 'value' )
```
